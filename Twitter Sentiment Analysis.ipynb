# Twitter Sentiment Analysis

```python
# Regex library
import re

# Read from the positive word file and make it into a list.
positive = open('positive.txt','r')
positive_data = positive.readlines()

# Remove new line(/n) from the data list.
positive_list = [i.strip() for i in positive_data]
print(positive_list)
```

```python
# Similarly fetch the negative words from the file and make it into a list.
negative = open('negative.txt','r')
negative_data = negative.readlines()

# Remove new line (/n) from the list
negative_list = [i.strip() for i in negative_data]
print(negative_list)
```

```python
# Get the stop words from the file and make it into a list.
stopwords= open('stopwords.txt','r')
stopwords_data = stopwords.readlines()

# Remove new line(/n) from the list.
stopwords_list = [i.strip() for i in stopwords_data]
print(stopwords_list)
```

```python
# Check if there is any common terms in the negative or positive list. Expecting no values to be returned
intersection_postive_negative_word = set(positive_list) & set(negative_list)
print(intersection_postive_negative_word)
```

```python
# Check if there is any common terms in the positive or stop_word list. Expecting no values to be returned
intersection_postive_stop_word = set(positive_list) & set(stopwords_list)
print(intersection_postive_stop_word)
```

```python
# Check if there is any common terms in the negative or stop_word list. Expecting no values to be returned
intersection_negative_stop_word = set(negative_list) & set(stopwords_list)
print(intersection_negative_stop_word)
```

## Reading tweets with #Trump

```python
# Read the file in UTF-8 format.
#Cleaning the dataset
with open("Trump.txt",'r',encoding='utf-8') as file:
    tweets = file.readlines()
    for item in tweets:
     # Removing the URLs, unwanted symbols and user names using regex and convert it to list format
     tweet_lowcase = item.lower()
     cleaning_data = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet_lowcase).split())
     # Remove the emojis by comparing the character and checking if it is a valid digit.
     remove_emoji = ' '.join(s for s in cleaning_data.split() if not any(c.isdigit() for c in s))
     # Convert the formatted string to list.
     tweet_list = list(remove_emoji.split(" "))
    print(tweet_list)
```

Count the number of words in the list. Once we do that, let us check the validity of positive and negative words by checking if both of them any common words. This is to ensure that we don't treat a single word in both positive and negative sentiment.

```python
# Count the total number of words.
count_total_words = len(tweet_list)
print('Total word count for #trump tweets:',count_total_words)
```

Using the lists that we created, compare it with the words in the tweet list and measure the sentiment.

```python
# Initialize the counts.
positive_count =0
negative_count =0
stopword_count =0
other_word_count =0

# For each word, increment the count based on the list.
for word in tweet_list:
    if word in positive_list:
        positive_count+=1
    elif word in negative_list:
        negative_count+=1
    elif word in stopwords_list:
        stopword_count+=1
    else:
        other_word_count+=1
print('The positive count of words in the tweets list:',positive_count)
print('The negative count of words in the tweets list:',negative_count)
print('The stopword count of words in the tweets list:',stopword_count)
print('Words not in positive/negative/stop words:',other_word_count)
```

Based on the count, measure the ratio of positive and negative words using the total word count.

```python
# Positive ratio and percentage.
ratio_of_positive = positive_count/count_total_words
print('Positive ratio', round(ratio_of_positive, 3))
Percentage_ratio_of_positive = ratio_of_positive * 100
print('Positive ratio percentage: ', round(Percentage_ratio_of_positive, 2),'%')
```

```python
# Negative ratio and percentage.
ratio_of_negative = negative_count/count_total_words
print('Negative ratio', round(ratio_of_negative, 3))
Percentage_ratio_of_negative = ratio_of_negative * 100
print('Negative ratio percentage: ', round(Percentage_ratio_of_negative, 2),'%')
```

```python
sentiment_score_trump = positive_count - negative_count
print(sentiment_score_trump)
if sentiment_score_trump > 0:
    print("Overall sentiment is positive")
elif sentiment_score_trump < 0:
    print("Overall sentiment is negative")
elif sentiment_score_trump == 0:
     print("Overall sentiment is neutral")
```

## Conclusion:

Based on the above calculation, it is clear that the overall sentiment is positive but it is only mildly positive since the negative ratio percentage is 3.36% vs the positive ratio percentage which is 5.95%.
